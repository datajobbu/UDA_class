```python
from konlpy.tag import Kkma
from gensim import models
from gensim import corpora
import numpy as np
import pandas as pd
import pyLDAvis.gensim
import gensim
```

```python
kor_corpus = pd.read_csv('C:\\Users\\user\\downloads\\news_koreanwave.csv', encoding='utf-8')
```

```python
contents = kor_corpus['contents']
```

```python
k = Kkma()
stopword = ['것','수','저']

pos = lambda d: ['/'.join(p) for p in k.pos(d) if ((p[1]=='NNG') & (p[0] not in stopword))]
text_co = [pos(doc) for doc in contents[:1000]]
```

```python
dictionary_ko = corpora.Dictionary(text_co)
dictionary_ko.save('ko.dict')
```

```python
print(text_co[0])
```

```python
tf_ko = [dictionary_ko.doc2bow(text) for text in text_co]
tfidf_model_ko = models.TfidfModel(tf_ko)
tfidf_ko = tfidf_model_ko[tf_ko]
corpora.MmCorpus.serialize('ko.mm', tfidf_ko)
```

```python
print(tfidf_ko.corpus[1])
print("-------------------------------------------------------------")
print(sorted(tfidf_ko.corpus[1], key=lambda x: x[1], reverse=True)[:10])
print("-------------------------------------------------------------")
print("136: ", dictionary_ko.get(136))
print("102: ", dictionary_ko.get(102))
print("-------------------------------------------------------------")
print(len(dictionary_ko))
```

## Train LDA

```python
ntopics, nwords = 10, 4
np.random.seed(333)

lda_ko = models.ldamodel.LdaModel(tfidf_ko, id2word=dictionary_ko, num_topics=ntopics)
topics = lda_ko.print_topics(num_words=nwords)
for topic in topics:
    print(topic)
```

```python
bow = tfidf_model_ko[dictionary_ko.doc2bow(text_co[0])]
sorted(lda_ko[bow], key=lambda x: x[1], reverse=True)
```

## LDA Visualize

```python
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_ko, tf_ko, dictionary_ko)
pyLDAvis.display(vis)
```
